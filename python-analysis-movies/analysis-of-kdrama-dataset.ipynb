```python
import matplotlib.pyplot as plt
df['Rating'].plot(kind='hist', bins=20)
plt.xlabel('Rating')
plt.title('Distribution of Ratings')
plt.show()
```

```python
top_20_movies = df.sort_values(by='Rating', ascending=False).head(20)

# Plotting
plt.figure(figsize=(10, 8))
plt.barh(top_20_movies['Name'], top_20_movies['Rating'], color='green')
plt.xlabel('Rating')
plt.ylabel('Movie Name')
plt.title('Top 20 Movies with Highest Ratings')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest rated at the top
for index, value in enumerate(top_20_movies['Rating']):
    plt.text(value, index, f"{value:.1f}", va='center')
plt.show()
```

```python
top_20_movies = df.nlargest(20, 'Rating')

plt.figure(figsize=(12, 6))
plt.scatter(top_20_movies['Duration'], top_20_movies['Rating'], color='blue', alpha=0.7)

# Add titles and labels
plt.title('Top 20 Movies: Duration vs Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')

# Optionally, add a grid and an average rating line
plt.axhline(y=top_20_movies['Rating'].mean(), color='red', linestyle='--', label='Average Rating')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python

top_20_movies = df.nlargest(10, 'Rating')

plt.figure(figsize=(12, 6))
plt.scatter(top_20_movies['Number of Episodes'], top_20_movies['Rating'], color='blue', alpha=0.7)

# Add titles and labels
plt.title('Top 20 Movies: Number of Episodes vs Rating')
plt.xlabel('Episodes')
plt.ylabel('Rating')

plt.xlim(0, 30)  # Adjust these limits based on your data

# Set ticks for axes in increments of 10 for x-axis and 0.5 for y-axis
plt.xticks(np.arange(0, 31, 5))  # X-axis ticks from 60 to 180 in increments of 10

# Optionally, add a grid and an average rating line
plt.axhline(y=top_20_movies['Rating'].mean(), color='red', linestyle='--', label='Average Rating')
plt.legend()
plt.grid(True)
annotated_episode_counts = set()

for index, row in top_20_movies.iterrows():
    episode_count = row['Number of Episodes']
    
    # Only annotate if this episode count hasn't been annotated yet
    if episode_count not in annotated_episode_counts:
        plt.text(episode_count + 0.2, row['Rating'] + 0.1, str(episode_count), fontsize=9, ha='left', va='bottom')
        annotated_episode_counts.add(episode_count)  # Add the episode count to the set

plt.show()
```

```python
df['Cast'] = df['Cast'].apply(lambda x: x.split(', ') if isinstance(x, str) else x)

# Explode the cast column to create a new row for each cast member
exploded_df = df.explode('Cast')

# Calculate the average rating for each cast member
average_ratings = exploded_df.groupby('Cast')['Rating'].mean().reset_index()

# Sort the DataFrame by average rating
average_ratings = average_ratings.nlargest(20, 'Rating')  # Get top 20 cast members by rating

# Create a bar plot for average ratings by cast member
plt.figure(figsize=(12, 6))
plt.barh(average_ratings['Cast'], average_ratings['Rating'], color='purple', alpha=0.7)

# Add titles and labels
plt.title('Top 20 Cast Members by Average Movie Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cast Members')

for index, value in enumerate(top_20_movies['Rating']):
    plt.text(value, index, f"{value:.1f}", va='center')
# Show the plot
plt.show()
```

```python
df['Genre'] = df['Genre'].apply (lambda x: x.split(', ') if isinstance(x, str) else x)
exploded_df = df.explode('Genre')
average_ratings = exploded_df.groupby('Genre')['Rating'].mean().reset_index()

# Sort the DataFrame by average rating
average_ratings = average_ratings.nlargest(10, 'Rating')  # Get top 20 cast members by rating

# Create a bar plot for average ratings by cast member
plt.figure(figsize=(12, 6))
plt.barh(average_ratings['Genre'], average_ratings['Rating'], color='purple', alpha=0.7)

# Add titles and labels
plt.title('Top 10 Genre by Average Movie Rating')
plt.xlabel('Average Rating')
plt.ylabel('Genre')

for index, value in enumerate(top_20_movies['Rating']):
    plt.text(value, index, f"{value:.1f}", va='center')
# Show the plot
plt.show()
```
